# Selecting the smallest of the largest

```python
df=pd.read_csv('C:/Users/WAJID KHAN/Downloads/pandas cookbook code/PandasCookbook_Code/data/movie.csv')
```

```python
filt=df[['movie_title','imdb_score','budget']]
filt.head()
```

```python
filt.nlargest(100,'imdb_score').head()
```

```python
filt.nlargest(100,'imdb_score').nsmallest(5,'budget')
```

# Selecting the largest of each group by sorting

```python
fil=df[['movie_title','title_year','budget','imdb_score']]
```
